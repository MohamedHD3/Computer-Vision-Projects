```python
import numpy as np
import cv2

def nothing(x):
    print(x)
    
cv2.namedWindow('frame')
cv2.createTrackbar('H', 'frame', 0, 255, nothing)
cv2.createTrackbar('S', 'frame', 0, 255, nothing)
cv2.createTrackbar('V', 'frame', 0, 255, nothing)

img_hsv = np.zeros((350,500,3), np.uint8)

while(1):
    h = cv2.getTrackbarPos('H', 'frame')
    s = cv2.getTrackbarPos('S', 'frame')
    v = cv2.getTrackbarPos('V', 'frame')
    
    img_hsv[:] = (h, s, v)
    img_bgr = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)
    
    cv2.imshow('frame', img_bgr)
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
# print(cap.get(3))
# print(cap.get(4))

while(1):
    _, frame = cap.read()
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    height, width, channels = frame.shape # return (height, width, 3(BGR))
    
    cx = int(width/2) # center of x
    cy = int(height/2) # center of y
    
    pixel_center = hsv_frame[cy, cx] # return (hue, saturation, value)
    hue_value = pixel_center[0]
    
    color = 'Undefind'
    if hue_value < 5 or 165 < hue_value < 190 :
        color = 'Red'
    elif hue_value < 22 or 190 < hue_value < 201:
        color = 'Orange'
    elif hue_value < 33 or 201 < hue_value < 213:
        color = 'Yellow'
    elif hue_value < 84 or 213 < hue_value < 255:
        color = 'Green'
    elif hue_value < 129:
        color = 'Blue'
    elif hue_value < 141:
        color = 'Violet'
    elif hue_value < 165:
        color = 'Pink'
    else:
        color = 'Undefind'
        
    pixel_center_bgr = frame[cy, cx] # return (blue, green, red)
    b, g, r = int(pixel_center_bgr[0]), int(pixel_center_bgr[1]), int(pixel_center_bgr[2])
    cv2.rectangle(frame, (cx-120, 7), (cx+140, 77), (255,255,255), -1)
    cv2.putText(frame, color, (cx-110, 57), 0, 2, (b,g,r), 3)
    cv2.circle(frame, (cx,cy), 4, (25,25,25), 3)
    
    cv2.imshow('frame', frame)
    k = cv2.waitKey(1)
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
